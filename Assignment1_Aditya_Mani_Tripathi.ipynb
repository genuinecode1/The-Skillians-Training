```python
rows = int(input("Rows: "))
for i in range(1,int(rows/2)):
    print(" "*(rows-i) + "*"*(2*i-1))
for i in range(rows - int(rows/2),0,-1):
    print(" "*(rows-i) + "*"*(2*i-1))

```
