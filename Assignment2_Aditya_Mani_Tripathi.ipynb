```python
#question no.20 leetcode
class Solution:
    def matches(self, top, ch):
        if ch == ')' and top == '(':
            return True
        elif ch == '}' and top == '{':
            return True
        elif ch == ']' and top == '[':
            return True
        else:
            return False

    def isValid(self, expression: str) -> bool:
        stack = []

        for ch in expression:
            # if opening bracket, push to stack
            if ch == '(' or ch == '[' or ch == '{':
                stack.append(ch)
            else:
                # closing bracket: check stack top
                if stack:
                    top = stack[-1]
                    if self.matches(top, ch):
                        stack.pop()
                    else:
                        return False
                else:
                    return False

        return len(stack) == 0
```

```python
#question no 682 leetcode
class Solution:
    def calPoints(self, operations):
        ans = []

        for op in operations:
            if op == "C":
                ans.pop()
                continue

            if op == "D":
                a = ans[-1]
                ans.append(a * 2)
                continue

            if op == "+":
                a = ans[-1]
                ans.pop()
                b = ans[-1]
                sum_val = a + b
                ans.append(a)
                ans.append(sum_val)
                continue

            # numeric value
            ans.append(int(op))

        total = 0
        for val in ans:
            total += val

        return total
```

```python
#question no. 2000 leetcode
class Solution:
    def reversePrefix(self, word: str, ch: str) -> str:
        if len(word) == 0:
            return word

        count = -1
        flag = True

        # find first occurrence of ch
        for c in word:
            if ch == c:
                count += 1
                flag = False
                break
            else:
                count += 1

        # if ch not found, return original word
        if flag:
            return word

        # reverse from index 0 to count
        word_list = list(word)
        i = 0
        while i < count:
            word_list[i], word_list[count] = word_list[count], word_list[i]
            i += 1
            count -= 1

        return "".join(word_list)
```

```python
#question no.1021
class Solution:
    def removeOuterParentheses(self, S: str) -> str:
        res, opened = [], 0
        for c in S:
            if c == '(' and opened > 0: res.append(c)
            if c == ')' and opened > 1: res.append(c)
            opened += 1 if c == '(' else -1

        return "".join(res)
```
